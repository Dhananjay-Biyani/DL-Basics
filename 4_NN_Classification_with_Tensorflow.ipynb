### Creating our own data for classification

```python
from sklearn.datasets import make_circles

# making 1000 sampels
n_samples =  1000

# create circles
x ,y = make_circles(n_samples,noise=0.03,random_state=42) #make a large circle containing a smaller circle
```

```python
# checking feature
x
```

```python
# checking label
y
```

### converting data into dataframe

```python
import pandas as pd
circle = pd.DataFrame({"X0":x[:,0] ,"X1":x[:,1] , "label":y})
```
